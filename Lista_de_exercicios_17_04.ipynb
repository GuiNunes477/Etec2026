```python
# 1. Par ou ímpar
# Entrada
numero = float(input("Me diga um número: "))
# Processamenti/saída
if(numero % 2 == 0):
  print(f"O número {numero} é par")
else:
  print(f"O número {numero} é ímpar")
```

```python
# 2. maior entre dois números
#entrada
numero1 = float(input("Me diga um número: "))
numero2 =  float(input("Me diga outro número: "))
#procesamento/saída
if(numero1 > numero2):
  print(f"O número {numero1} é maior que o {numero2}")
elif(numero1 == numero2):
  print("Os números são iguais")
else:
  print(f"o {numero1} é menor que o {numero2}")
```

```python
# 3. Aprovação do aluno
#Entrada
nota1 = float(input("Insira sua 1º nota "))
nota2 = float(input("Insira sua 2º nota "))
nota3 = float(input("Insira sua 3º nota "))
nota4 = float(input("Insira sua 4º nota "))
# Processamento
media = (nota1+nota2+nota3+nota4)/4
#saída
if(media >= 7 and media <=10):
  print(f"Parabéns você passou, sua media foi {round(media,2)}")
elif(media >= 5 and media < 7):
 print(f"Recuperação sua média é {round(media,2)}")
elif(media < 5):
 print(f"Você foi reprovado sua média foi {round(media,2)}")
else:
 print("nota invalida")
```

```python
# 4. classificação de idade
#entrada
idade = int(input("Qual a sua idade? "))
# 3. processamento/saída
if(idade <=12 and idade >= 0):
  print(f"Você é uma criança, tem {idade} anos")
elif(idade > 12 and idade <= 17):
  print(f"Você é um adolescente, tem {idade} anos")
elif(idade > 17 and idade <=59 ):
  print(f"Você é um adulto, tem {idade} anos")
else:
  print(f"Você é idoso, tem {idade} anos")
```

```python
# 6. positivo, negativo ou zero
#entrada
num = float(input("Me fale um número: "))
#processamento/saída
if(num > 0):
  print(f"O número {num} é positivo")
elif(num == 0):
  print(f"O número é {num}")
else:
  print(f"O número {num} é negativo")
```

```python
# 7. escrever uma calculadora
#entrada
print("soma = +")
print("subtração = -")
print("multiplicação = *")
print("divisão = /")
nm1 = float(input("Insira um número " ))
operacao = input("Escolha uma operação ")
nm2 = float(input("Insira outro número "))
#processamento/saída
if (operacao  == "+"):
 print(f"A soma de {nm1} com {nm2} é {round(nm1 + nm2,2)}")
elif (operacao  == "-"):
 print(f"A subtração de {nm1} com {nm2} é {round(nm1 - nm2,2)}")
elif (operacao  == "*"):
 print(f"A multiplicação de {nm1} com {nm2} é {round(nm1 * nm2,2)}")
elif (operacao  == "/"):
 print(f"A divisão de {nm1} com {nm2} é {round(nm1 / nm2,2)}")
else:
  print("Opção invalida")
```

```python
# 8. Dia da semana
#entrada
dia = int(input("insira um núemro de 1 a 7: "))
#processamneto/saída
if(dia == 1):
  print("Hoje é domingo")
elif(dia == 2):
  print("Hoje é segunda")
elif(dia == 3):
  print("Hoje é Terça")
elif(dia == 4):
  print("Hoje é Quarta")
elif(dia == 5):
  print("Hoje é Quinta")
elif(dia == 6):
  print("Hoje é Sexta")
elif(dia == 7):
  print("Hoje é Sabádo")
else:
  print("Opção invalida")
```

```python
# 9. Ano bissexto
#Entrada
ano = int(input("Insira um ano: "))
#processamento/saída
if(ano % 4 == 0):
  print(f"o ano de {ano} é bissexto")
elif(ano % 100 == 0):
  print(f"o ano de {ano} não é bissexto")
elif(ano % 400 == 0):
  print(f"O ano de {ano} é bissexto")
else:
  print(f"O ano de {ano} não é bissexto")
```

```python
# 10. Desconto em compras
#entrada
valor = float(input("Qual o valor da sua compra? "))
#processamento/saída
if(valor < 50 and valor > 0):
  print(f"Você não tem desconto pagará R${valor}")
elif(valor >= 50 and valor < 100):
  print(f"Você tem um desconto de 5% pagará R${round(valor - (valor * 0.05),2)}")
elif(valor < 0):
  print("Invalida")
else:
  print(f"Você tem um desconto de 10% pagará R${round(valor - (valor* 0.10),2)} ")
```
