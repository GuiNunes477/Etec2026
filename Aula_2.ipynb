Atividade: Fazer um algoritmo que depois recebe 4 notas do aluno e calcula a média aritimética e imprime a média





Algoritimo

#Entrada

*   ler nota1
*   ler nota2
*   ler nota3
*   ler nota4




```python
#Entrada
nota1 = float(input("Qual a primeira nota? "))
nota2 = float(input("Qual a segunda nota? "))
nota3 = float(input("Qual a terceira nota? "))
nota4 = float(input("Qual a quarta nota? "))
#Processamento
media = ((nota1 + nota2 + nota3 + nota4)/ 4)
#Saída
if(media >= 6):
  print("Parabéns Você passou sua média é", round(media,2))
elif(media >= 5):
  print("Passou por pouco, se esforçe mais sua média foi", round(media,2))
else:
  print("Você reprovou volte ano que vem sua média é", round(media,2))
```
