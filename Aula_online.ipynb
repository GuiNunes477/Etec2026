```python
# 1. soma simples
#entrada
numero1 = int(input("Insira um número inteiro "))
numero2 = int(input("Insira outro número inteiro "))

#processsmento
soma = numero1+numero2

#saída
print(f"A soma de {numero1} com {numero2} é: {soma}")
```

```python
# 2. subtrção
#entrada
numero3 = int(input("Insira um número inteiro "))
numero4 = int(input("Insira outro número inteiro "))

#processsmento
subtração = numero3 - numero4

#saída
print(f"A subtração de {numero3} com {numero4} é: {subtração}")
```

```python
# 3. multiplicação
#entrada
numero5 = float(input("Insira um número "))
numero6 = float(input("Insira outro número "))

#processsmento
multi = numero5 * numero6

#saída
print(f"A multiplicação de {numero5} com {numero6} é: {multi}")
```

```python
# 4. divisão
#entrada
numero7 = float(input("Insira um número "))
numero8 = float(input("Insira outro número "))

#processsmento
divisao = numero7 /  numero8

#saída
print(f"A divisão de {numero7} com {numero8} é: {round(divisao,2)}")
```

```python
# 5. média
#entrada
numero9 = float(input("Insira a primeira nota "))
numero10 = float(input("Insira a segunda nota  "))
numero11 = float(input("Insira a terceira nota "))
#processsmento
media = (numero9 + numero10 + numero11) / 3

#saída
print(f"A Média das notas é: {round(media,2)}")
```

```python
# 6. divisão
#entrada
numero12 = int(input("Insira um número inteiro "))

#processsmento
dobro = numero12 * 2
metade = numero12 / 2
resto = numero12 % 3

#saída
print(f"O dobro de {numero12} é {dobro}, já a metade é {metade}")
print(f"O resto de {numero12} divididoa por 3 é {resto}")
```

```python
# 7. conversão de temperatura
#entrada
c = float(input("Qual a temperatura de hoje? "))
#Processamento
f = ((c*1.8)+32)
#Saida
print(f"Hoje está {c}º celsius que em fahrenheit são {f} fahrenheit")
```

```python
#8. salário
#entrada
salario = float(input("Qual o seu salário? "))
#Processamento
aumento = salario * 1.10
#saída
print(f"Com um aumento de 10% o seu salarío agora é {round(aumento,2)}")
```

```python
# 9. Area do retangulo

#entrada
base = float(input("Qual a base do retangulo? "))
altura = float(input("Qual a altura do retangulo? "))
#processamento
area = base * altura
#Saída
print(f"A área do seu retangulo é {round(area,2)}")
```

```python
# 10. consumo de combustivel

#entrada
distancia = float(input("Quantos km você rodou? "))
litros = float(input("Quantos litros você gastou? "))
#processamento
km_h = distancia/litros
#saída
print(f"Você fez {km_h} km/l")
```
