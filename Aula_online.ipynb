```python
# 1. soma simples
#entrada
numero1 = int(input("Insira um número inteiro "))
numero2 = int(input("Insira outro número inteiro "))

#processsmento
soma = numero1+numero2

#saída
print(f"A soma de {numero1} com {numero2} é: {soma}")
```

```python
# 2. subtrção
#entrada
numero3 = int(input("Insira um número inteiro "))
numero4 = int(input("Insira outro número inteiro "))

#processsmento
subtração = numero3 - numero4

#saída
print(f"A subtração de {numero3} com {numero4} é: {subtração}")
```

```python
# 3. multiplicação
#entrada
numero5 = float(input("Insira um número "))
numero6 = float(input("Insira outro número "))

#processsmento
multi = numero5 * numero6

#saída
print(f"A multiplicação de {numero5} com {numero6} é: {multi}")
```

```python
# 4. divisão
#entrada
numero7 = float(input("Insira um número "))
numero8 = float(input("Insira outro número "))

#processsmento
divisao = numero7 /  numero8

#saída
print(f"A divisão de {numero7} com {numero8} é: {round(divisao,2)}")
```

```python
# 5. média
#entrada
numero9 = float(input("Insira a primeira nota: "))
numero10 = float(input("Insira a segunda nota:  "))
numero11 = float(input("Insira a terceira nota: "))
#processsmento
media = (numero9 + numero10 + numero11) / 3

#saída
print(f"A Média das notas é: {round(media,2)}")
```

```python
# 6. divisão
#entrada
numero12 = int(input("Insira um número inteiro: "))

#processsmento
dobro = numero12 * 2
metade = numero12 / 2
resto = numero12 % 3

#saída
print(f"O dobro de {numero12} é {dobro}, já a metade é {metade}")
print(f"O resto de {numero12} divididoa por 3 é {resto}")
```

```python
# Estrutura de decisões(if, Else, Elif)

#entrada
a = int(input("Insira um número inteiro "))
b = int(input("Insira outro numero inteiro "))

#processamento

if(a == b):
  print(f"{a} é igual a {b}")
elif(a == 6 and b == 7):
  print(f"SIX SEVEN")
else:
  print(f"{a} é diferente de {b}")
```

```python
#verificar se o número é par ou ímpar

#entrada
n1 = int(input("Insira um número: "))
#Saída
if(n1 % 2 == 0):
  print(f"O número {n1} é par")
else:
  print(f"O número {n1} é impar")
```

```python
#Verificar idade

#entrada
nome = input("Qual o seu nome? ")
idade = int(input("Quantos anos você tem? "))

#processamento/saída
if(idade >= 12 and idade <= 18):
  print(f"{nome} Você é um adolescente")
elif(idade >= 18 and idade != 67):
  print(f"{nome} Você é um adulto")
elif(idade == 67):
  print("SIX SEVEN")
else:
  print(f"{nome}, Você é criança")
```

```python
#escrever uma calculadora
print("------------CALCULADORA------------")
print("                                   ")
print("---------------MENU----------------")
print("                                   ")
print("soma = +")
print("subtração = -")
print("multiplicação = *")
print("divisão = /")
print("                                   ")

nm1 = float(input("Insira um número " ))

operacao = input("Escolha uma operação ")

nm2 = float(input("Insira outro número "))

if (operacao  == "+"):
 print(f"A soma de {nm1} com {nm2} é {round(nm1 + nm2,2)}")

if (operacao  == "-"):
 print(f"A subtração de {nm1} com {nm2} é {round(nm1 - nm2,2)}")

if (operacao  == "*"):
 print(f"A multiplicação de {nm1} com {nm2} é {round(nm1 * nm2,2)}")

if (operacao  == "/"):
 print(f"A divisão de {nm1} com {nm2} é {round(nm1 / nm2,2)}")

else:
  print("Opção invalida")
```
