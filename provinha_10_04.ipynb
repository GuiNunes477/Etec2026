```python
# 2. Escrever um programa que verifica as notas de 4 bimestres, calcule a média dos alunos
# Não deixe a nota ser < que 0 ou > que 10
# < que 5 = reprovado
# > que 5 e < 7 = recuperação
# >= 7 = Aprovado
#Entrada
nota1 = float(input("Qual a sua nota do 1º bimestre? "))

nota2 = float(input("Qual a sua nota do 2º bimestre? "))

nota3 = float(input("Qual a sua nota do 3º bimestre? "))

nota4 = float(input("Qual a sua nota do 4º bimestre? "))

#Processamneto
media = (nota1 + nota2 + nota3 + nota4)/4
#Saída
if(media < 5 and media> 0):
  print(f"sua média é {round(media,2)} você está Reprovado(a)")

elif(media >10):
  print("Nota invalida")

elif(media <0):
  print("Nota invalida")

elif(media < 7 and media > 5):
  print(f"Sua média é {round(media,2)} Você está de recuperação ")

else:
  print(f"Sua média é {round(media,2)} Parabens você foi aprovado(a)")
```
